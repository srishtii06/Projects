```python
import pandas as pd
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

RESUME PROJECTS-Multiple Disease Prediction Model

```python
disease=pd.read_csv('https://raw.githubusercontent.com/YBIFoundation/Dataset/main/MultipleDiseasePrediction.csv')
```

```python
disease['prognosis'].value_counts()
```

```python
y=disease['prognosis']
```

```python
x=disease.drop(['prognosis'],axis=1)
```

```python
#x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=2529)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,stratify=y,random_state=2529)
```

120 samples for each disease ,so 75% of each 120 should go to train so 90 must be in train and rest 30 in test ;because training data should be more; using stratify=y ;the stratify parameter in train_test_split() splits the dataset in a way that ensures the proportion of values in the sample is the same as the proportion provided with this parameter

```python
y_train.value_counts()
```

```python
from sklearn.neighbors import KNeighborsClassifier # not logistic as used for binary outputs
```

```python
model=KNeighborsClassifier()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
print(confusion_matrix(y_test,y_pred))
```

```python
print(classification_report(y_test,y_pred))
```

```python
from sklearn.tree import DecisionTreeClassifier,plot_tree
```

```python
model=DecisionTreeClassifier(max_depth=5)
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,y_pred))
```

```python
fig = plt.subplots(figsize=(15,10))
plot_tree(model, feature_names=x.columns,filled=True);
```
