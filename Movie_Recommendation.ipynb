```python
x=df[[ 'Movie_Genre',  'Movie_Keywords', 'Movie_Tagline', 'Movie_Cast',
       'Movie_Director']]
```

```python
x.isna().sum()
```

```python
x.info()
```

```python
x.fillna('',inplace=True)
```

```python
x.isna().sum()
```

```python
x.shape
```

```python
features=x['Movie_Genre']+' '+x['Movie_Keywords']+' '+x['Movie_Tagline']+' '+x['Movie_Cast']+' '+x[ 'Movie_Director']
```

```python
features.shape
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf=TfidfVectorizer()
```

```python
x_movie=tfidf.fit_transform(features)
```

```python
print(x_movie)
```

```python
#cal distance
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score=cosine_similarity(x_movie)
```
